* 다음 두 가지 산출물은 서로 다른 두 가지 메커니즘을 사용하여 제품 연관성을 나타낸다. 
* 마지막 두 가지는 이 데이터 집합에서 성별을 사용하는 것이 현명한지 여부를 탐구하기 위한 인구통계학적 분석을 나타낸다.

--------

* Mean Rating : 각 영화의 평균 등급을 계산하고 가장 높은 등급이 먼저 나열된 순서대로 상위 3개(상위 2의 평균 점수와 함께)를 제출한다.
* Rating Count(popularity) : 각 영화의 등급의 개수를 구하고, 가장 많은 등급의 개수를 먼저 추천하고, 상위 3개(상위 2개 카운트와 함께)를 제출한다.
* % of ratings 4+ (liking) : 등급이 4이상인 영화의 등급 비율을 계산한다. 가장 높은 등급 비율을 가진 것을 먼저 추천해주고, 상위 3개(상위 2개 백분율과 함께) 제출한다.
* 토이 스토리를 본 사람들의 인기 영화 : 수업에서 설명 된 (x + y) / x 방법을 사용하여 영화 "토이 스토리"에서 가장 자주 발생하는 영화를 계산하십시오. 즉, 각 영화에 대해 해당 영화를 평가 한 토이 스토리 평가자의 비율을 계산하십시오. 가장 높은 비율로 먼저 주문하고 상위 3 개 (상위 2 개에 대한 상관 관계와 함께)를 제출하십시오. 유대 관계를 가지십시오. 유대 관계를 끊으려면 가장 낮은 번호의 영화를 가장 높은 순위의 영화로 사용하십시오. 즉, 영화 541과 318이 묶여 있으면 318이 더 높은 순위를 얻습니다.  
* 토이 스토리와의 상관관계 : 토이 스토리의 평점 벡터와 서로의 영화 간의 상관 관계를 계산합니다. 가장 높은 상관 관계 (양수 만)로 정렬하고 상위 2에 대한 상관 값과 함께 상위 3 개를 제출하십시오. 동시 발생과 상관 관계의 차이점을 확인하십시오. 이러한 측정 항목은 다양한 유형의 관계를 보여줍니다.
* 성별에 따른 평균 평점 차이 : 먼저 남성과 여성에 대해 각 영화의 평균 등급을 개별적으로 다시 계산하십시오. 그리고 남성과 여성에 대한 전체 평균 평점 (모든 평점에 대한)을 계산합니다. 평균 차이와 함께 가장 큰 차이가있는 두 영화 (남자는 여자보다 높은 영화, 여자는 남자보다 높은 영화)를 제출합니다. 또한 전체 평점 평균 (여성 평균-남성 평균)의 차이를 제출하십시오.
* 성별에 따른 % of ratings 4+ : 위와 같이 두 개의 영화를 제출하십시오. (각 방향에서 가장 큰 차이). 그리고 다시 남성이나 여성이 별 4 개 이상의 영화를 평가할 가능성이 더 높은지를 나타냅니다.

## Load data

```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("HW1-data.csv", index_col="User")
data.head()
```

## 1. Mean Rating

```python
movie = data.T #뒤집기
gender = movie .loc["Gender (1 =F, 0=M)"] #남녀 계산을 편하게 하기 위해 잠시 지우자
movie.drop("Gender (1 =F, 0=M)", inplace = True)
calculator = movie .copy() #계산을 위한 데이터 프레임
movie.head()
```

```python
mean = (calculator.sum(axis =1)/calculator.count(axis =1))
movie["Mean"] = mean
movie.head()
```

Shawshank Redemption, The (1994), Star Wars: Episode IV - A New Hope (1977), Blade Runner (1982) 가 상위 세개이다.

```python
mean.sort_values(ascending = False)
```

## 2. Rating Count(popularity)

```python
Popularity = calculator.count(axis =1)
movie["Popularity"] = Popularity
movie.head()
```

```python
Popularity.sort_values(ascending = False)
```

## 3. % of ratings 4+ (liking)

```python
rating = (calculator>=4).astype(int) #평점이 4이상인 것을 1로 바꾸기
rating.head()
```

```python
plus_4 = rating.sum(axis=1)
plus_4
```

```python
Positive_rating = plus_4/Popularity #평점이 4이상인 비율 구하기
movie["Positive_rating"] = Positive_rating
movie.head()
```

```python
Positive_rating.sort_values(ascending = False)
```

## 4. Association with Toy Story

```python
movie.shape
```

```python
association = []
for i in range(20):
    a = calculator.iloc(0)[i].notnull().values # 비교할 영화
    toy_story = calculator.loc["1: Toy Story (1995)"].notnull().values
    inter = a*toy_story # and 연산
    inter = list(inter)
    toy_story = list(toy_story)
    association.append(inter.count(True)/toy_story.count(True)) # x+y/x
movie["Association with Toy Story"] = association
movie.head()
```

```python
# 내림차순 정렬 (맨 앞은 toy story이니 무시)
association.sort(reverse=True)
association
```

## 5. Ratings correlation with Toy Story

```python
corr = data.corr(method = 'pearson') #상관계수 분석
toy_cor = list(corr["1: Toy Story (1995)"])
del toy_cor[0] #첫번째 요소는 성별이니 삭제한다.
movie["Correlation with Toy Story"] = toy_cor
movie.head()
```

## 6. Mean rating difference by gender

```python
female = data[data["Gender (1 =F, 0=M)"] == 1].T
male = data[data["Gender (1 =F, 0=M)"] == 0].T

female.drop("Gender (1 =F, 0=M)", inplace = True)
male.drop("Gender (1 =F, 0=M)", inplace = True)
```

```python
female_mean = (female.sum(axis =1)/female.count(axis =1))
male_mean = (male.sum(axis =1)/male.count(axis =1))

movie["Female_mean"] = female_mean
movie["Male_mean"] = male_mean
movie["mean differences"] = movie["Female_mean"] - movie["Male_mean"]
movie.head()
```

```python
# 음수는 남자가 평균이 더 큰 경우, 양수는 여자가 평균이 더 큰 경우이다.
movie["mean differences"].sort_values()
```

## 7. % of ratings 4+ separately for males and females

```python
female_rating = (female>=4).astype(int) #평점이 4이상인 것을 1로 바꾸기(여자)
male_rating = (male>=4).astype(int) #평점이 4이상인 것을 1로 바꾸기(남자)

female_plus_4 = female_rating.sum(axis=1)
male_plus_4 = male_rating.sum(axis=1)
```

```python
female_popularity = female.count(axis =1)
male_popularity = male.count(axis =1)

Female_positive_rating = female_plus_4/female_popularity #평점이 4이상인 비율 구하기
Male_positive_rating = male_plus_4/male_popularity #평점이 4이상인 비율 구하기

movie["Female_positive_rating"] = Female_positive_rating
movie["Male_positive_rating"] = Male_positive_rating
movie["Positive_difference"] = movie["Female_positive_rating"] - movie["Male_positive_rating"]
movie.head()
```

```python
# 음수는 남자가 평균이 더 큰 경우, 양수는 여자가 평균이 더 큰 경우이다.
movie["Positive_difference"].sort_values()
```

```python
import surprise
#데이터를 읽어들이는 객체 생성(rating scale : 평점의 범위)
reader = surprise.Reader(rating_scale = (1,5))
reader
```

```python
data.head()
```

```python
recommend = data.drop("Gender (1 =F, 0=M)", axis=1)
```

```python
recommend = recommend.stack()
recommend_df = recommend.reset_index()
col = ["User", "Movie", "Rating"]
recommend_df.columns = col
recommend_df.head()
```

```python
# "User", "Movie", "Rating" 순으로 데이터 생성
data = surprise.Dataset.load_from_df(recommend_df[col], reader)
```

```python
print(col)
print(data)
```

```python
# 학습
trainset = data.build_full_trainset()
option = {'name' : 'pearson'} #pearson similarity 사용
algo = surprise.KNNBasic(sim_options=option) #surprise.KNNBasic 사용
algo.fit(trainset) #학습
```

```python
User_list = list(set(recommend_df.User))
Movie_list = list(set(recommend_df.Movie))
```

```python
who = int(input("ID를 입력하시오: "))

index = User_list.index(who)
print('user_index: ', index)

result = algo.get_neighbors(index, k=3) #knn 모델, 유사한 사용자 3명을 보이게 함ㄴ
print('당신과 유사한 고객은?: ', result)

print('당신에게 추천하는 상품: ', '\n')

recommend_item = [] #추천할 item
for r1 in result:
    max_rating = data.df[data.df['User']==User_list[r1]]['Rating'].max()
    cos_id = data.df[(data.df['Rating']==max_rating)&(data.df['User']==User_list[r1])]['Movie'].values
    recommend_item.extend(cos_id)
                     
print(recommend_item)
```
